# Dataset

Lista de puntajes obtenidos por diferentes jugadores oficialmente reconocidos por la Federación Internacional de Ajedrez. Los puntajes son mensuales y se cuenta con información de todos los meses del 2025. Se han considerado solo los puntajes del modo de juego "Blitz". Se descargaron 12 archivos XML con los puntajes, que luego fueron integrados en 1 solo archivo.

Fuente: https://ratings.fide.com/download_lists.phtml

# Preprocesamiento

Se aplicaron cinco decisiones de limpieza sobre el dataset original:

1. Eliminar columna `Unnamed: 0`. Columna residual generada al exportar el CSV desde pandas. Duplica el índice sin aportar información.
2. Consolidar columnas de título. Se fusionan `title` y `w_title` en una sola columna `title`, tomando `title` como valor primario y `w_title` como alternativa cuando `title` es nulo. `o_title` (árbitros, entrenadores) y `foa_title` (arena online) corresponden a categorías distintas a los títulos de juego y no son relevantes para el análisis de ratings, por lo que se eliminan.
3. Extraer mes desde `filename`. La columna `filename` codifica el mes en formato `blitz_jan25frl_xml.xml`. Se extrae como entero (1–12) en una nueva columna `month` y se elimina `filename`. Sin esta variable el análisis temporal no es viable.
4. Recodificar `flag` como columna `active`. `flag` codifica simultáneamente el estado de actividad y el sexo del jugador (`i`=inactivo masculino, `w`=femenino activa, `wi`=femenino inactiva, NaN=masculino activo). Se extrae únicamente el componente de actividad como booleano (`active`=True si el jugador estuvo activo en ese período). La información de sexo ya está contenida en la columna `sex`, por lo que no se pierde información al eliminar `flag`.
5. Se mantienen las filas con valores perdidos en `birthday`, pero se renombra la columna a `birth_year`.

```python
print("Estado inicial:")
print(f"  Shape: {df.shape}")
print(f"  Columnas: {list(df.columns)}")
print()
print(df.head(3).to_string())
```

```python
month_map = {'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,
             'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}

# Paso 1: drop Unnamed: 0
df = df.drop(columns=['Unnamed: 0'])

# Paso 2: coalesce title columns, drop w_title / o_title / foa_title
df['title'] = df['title'].fillna(df['w_title'])
df = df.drop(columns=['w_title', 'o_title', 'foa_title'])

# Paso 3: extract month from filename, drop filename
df['month'] = df['filename'].str.extract(r'blitz_([a-z]+)25')[0].map(month_map)
df = df.drop(columns=['filename'])

# Paso 4: flag → active boolean, drop flag
# active=True: flag is NaN (active male) or 'w' (active female)
# active=False: flag is 'i' (inactive male) or 'wi' (inactive female)
df['active'] = ~df['flag'].isin(['i', 'wi'])
df = df.drop(columns=['flag'])

# Paso 5: rename birthday to birth_year
df = df.rename(columns={'birthday': 'birth_year'})
```

```python
print("Estado final:")
print(f"  Shape: {df.shape}")
print(f"  Columnas: {list(df.columns)}")
print()
print(df.head(3).to_string())
```

# Transformaciones

Se aplicaron dos transformaciones sobre el dataset preprocesado, ambas orientadas a habilitar el análisis exploratorio por grupos.

1. `age`: A partir de `birth_year` se calcula `age = 2025 - birth_year`. Los registros sin `birth_year` producen `age` nulo, lo que es aceptable: estos jugadores quedan fuera de los análisis que requieran esta variable.
2. `age_group`: Sobre `age` se aplica un binning en siete tramos: ≤14, 15-18, 19-25, 26-35, 36-50, 51-65 y 65+, para ser usados a modo de comparación.
3. `rating_category`: El `rating` Elo es continuo (1400–2800+). Se definen cinco niveles basados en umbrales del ajedrez competitivo: Principiante (<1500), Intermedio (1500–1800), Avanzado (1800–2100), Experto (2100–2400) y Élite (≥2400), para aplicar comparaciones.

```python
# T1: age (variable derivada)
df['age'] = (2025 - df['birth_year']).astype('Int64')

# T2: age_group (binning)
age_bins   = [0, 14, 18, 25, 35, 50, 65, 200]
age_labels = ['≤14', '15-18', '19-25', '26-35', '36-50', '51-65', '65+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

# T3: rating_category (binning)
rating_bins   = [0, 1500, 1800, 2100, 2400, 3000]
rating_labels = ['Principiante', 'Intermedio', 'Avanzado', 'Experto', 'Élite']
df['rating_category'] = pd.cut(df['rating'], bins=rating_bins, labels=rating_labels)
```

# Análisis Exploratorio de Datos

Se plantean cuatro hipótesis sobre el comportamiento esperado del dataset, evaluadas con tablas de resumen.

**H1**: La mayoría de los jugadores registrados están activos.

```python
# H1: La mayoría de los jugadores está activo
print(df['active'].value_counts()
      .to_frame('registros')
      .assign(porcentaje=lambda x: (x['registros'] / x['registros'].sum() * 100).round(1)))
```

**Rechazada**. Solo el 39.2% de los registros corresponde a jugadores activos; el 60.8% está marcado como inactivo por la FIDE.

**H2**: El rating promedio y mediano no varía a lo largo del año.

```python
# H2: El rating no varía a lo largo del año
print(df.groupby('month')['rating'].agg(media='mean', mediana='median').round(1))
```

**Rechazada**. La media cae de 1748 (enero) a 1733 (diciembre) y la mediana de 1725 a 1705 — un descenso sistemático de ~15 puntos. El patrón refleja la incorporación mensual de nuevos jugadores con ratings más bajos, no el deterioro de los ya registrados.

**H3**: La proporción de jugadores inactivos es mayor en los niveles de rating más bajos.

```python
# H3: Mayor proporción de inactivos en ratings bajos
print(df.groupby('rating_category', observed=True)['active']
      .agg(total='count',
           pct_activo=lambda x: round(x.mean() * 100, 1),
           pct_inactivo=lambda x: round((~x).mean() * 100, 1)))
```

**Rechazada**. La mayor tasa de inactividad se encuentra en el nivel Avanzado (64.7%), no en Principiante (54.7%). Los jugadores de Élite tienen la menor tasa (45.0%), y los de Principiante incluyen muchos nuevos jugadores activamente compitiendo para establecer su rating.

**H4**: Los jugadores de mayor edad tienen ratings más altos.

```python
# H4: Jugadores mayores tienen ratings más altos
print(df.groupby('age_group', observed=True)['rating']
      .agg(mediana='median', media='mean', n='count')
      .round(1))
```

**Parcialmente confirmada**. El rating mediano crece con la edad hasta el tramo 51-65 (mediana 1841), pero desciende en el grupo 65+ (1819). Los jugadores más jóvenes (≤14) tienen la mediana más baja (1531).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import Rectangle

# ============================================================
# 1. FILTRAR ULTIMO MES Y JUGADORES ACTIVOS
# ============================================================
#ultimo_mes = df["month"].max()
#df_dic = df[(df["month"] == ultimo_mes) & (df["active"] == True)].copy()

# ============================================================
# FILTRAR DICIEMBRE, SOLO JUGADORES ACTIVOS QUE YA ESTABAN EN ENERO
# ============================================================
mes_enero = 1
mes_diciembre = 12

# Jugadores activos registrados en enero
jugadores_enero_activos = df.loc[
    (df["month"] == mes_enero) &
    (df["active"] == True),
    "fideid"
].dropna().unique()

# Jugadores activos en diciembre que ya estaban activos en enero
df_dic = df[
    (df["month"] == mes_diciembre) &
    (df["active"] == True) &
    (df["fideid"].isin(jugadores_enero_activos))
].copy()

print(f"Jugadores activos en enero: {len(jugadores_enero_activos):,}")
print(f"Jugadores activos en diciembre y ya registrados activos en enero: {df_dic['fideid'].nunique():,}")

# ============================================================
# 2. RECLASIFICAR EN 3 CATEGORIAS DE RATING
# ============================================================
def clasificar_rating(r):
    if r < 1600:
        return "Rating < 1600"
    elif r < 2000:
        return "1600 - 1999"
    else:
        return "Rating >= 2000"

df_dic["categoria_rating"] = df_dic["rating"].apply(clasificar_rating)

orden_categorias = [
    "Rating < 1600",
    "1600 - 1999",
    "Rating >= 2000"
]

orden_k = [40, 20, 10]

# ============================================================
# 3. MATRIZ DE CONTEOS Y PORCENTAJES
# ============================================================
conteo_k = (
    df_dic.groupby(["categoria_rating", "k"], observed=True)["fideid"]
          .nunique()
          .reset_index(name="jugadores_activos")
)

matriz_conteo = (
    conteo_k
    .pivot(index="categoria_rating", columns="k", values="jugadores_activos")
    .reindex(index=orden_categorias, columns=orden_k)
    .fillna(0)
    .astype(int)
)

matriz_pct = matriz_conteo.div(matriz_conteo.sum(axis=1), axis=0).mul(100)

# ============================================================
# 3.1 FORMATEAR CONTEOS EN K
# ============================================================
def abreviar_n(n):
    if n >= 1000:
        valor = f"{n/1000:.1f}K"
        return valor.replace(".0K", "K")
    return str(n)

matriz_conteo_txt = matriz_conteo.map(abreviar_n)

annot = (
    matriz_conteo_txt +
    "\n(" +
    matriz_pct.round(1).astype(str) +
    "%)"
)

# ============================================================
# 4. GRAFICO
# ============================================================
fig, ax = plt.subplots(figsize=(11.5, 6.2))

COL_DARK = "#2D4356"
COL_LIGHT = "#F4EFE6"
COL_RED = "#C0392B"
COL_GOLD = "#C99700"

cmap_custom = LinearSegmentedColormap.from_list(
    "rating_k_style",
    [COL_LIGHT, "#9EB7C8", COL_DARK]
)

sns.heatmap(
    matriz_pct,
    annot=annot,
    fmt="",
    cmap=cmap_custom,
    linewidths=1.4,
    linecolor="white",
    cbar=False,
    annot_kws={
        "fontsize": 11,
        "fontweight": "600"
    },
    ax=ax
)

# ============================================================
# 5. FORMATO DE TEXTO
# ============================================================
ax.set_xlabel("Multiplicador de victorias y derrotas", fontsize=11, color="#333333", labelpad=12)
ax.set_ylabel("Nivel de juego", fontsize=11, color="#333333", labelpad=12)

nombres_y = [
    "Nivel < 1600",
    "1600 - 1999",
    "Nivel >= 2000"
]

nombres_x = [
    "Alto",
    "Medio",
    "Bajo"
]

ax.set_yticklabels(nombres_y, rotation=0, fontsize=10.5, color="#333333")
ax.set_xticklabels(nombres_x, rotation=0, fontsize=10, color="#333333")

# Ajustar color del texto dentro del heatmap según intensidad
for text in ax.texts:
    try:
        value = float(
            text.get_text()
            .split("\n")[1]
            .replace("(", "")
            .replace("%)", "")
        )
        text.set_color("white" if value >= 45 else "#333333")
        text.set_fontsize(11)
        text.set_fontweight("600")
    except Exception:
        pass

# ============================================================
# 6. RESALTAR CELDA CLAVE
# ============================================================
ax.add_patch(
    Rectangle(
        (0, 2), 1, 1,
        fill=False,
        edgecolor=COL_RED,
        linewidth=6
    )
)

# ============================================================
# 7. ANOTACION EXTERNA
# ============================================================
x_inicio_flecha = -0.09
y_inicio_flecha = -0.08

# Flecha diagonal corta hacia la esquina inferior izquierda del recuadro
ax.annotate(
    "",
    xy=(0.0, 3.0),                 # esquina inferior izquierda del recuadro
    xycoords="data",
    xytext=(x_inicio_flecha, y_inicio_flecha),
    textcoords="axes fraction",
    annotation_clip=False,
    arrowprops=dict(
        arrowstyle="->",
        color=COL_RED,
        lw=2.2,
        shrinkA=0,
        shrinkB=0
    ),
    clip_on=False
)

# Texto un poco más abajo
ax.text(
    x_inicio_flecha - 0.09, y_inicio_flecha - 0.06,
    "Aquí están las",
    transform=ax.transAxes,
    ha="left",
    va="bottom",
    fontsize=11,
    fontweight="600",
    color=COL_RED,
    clip_on=False
)

ax.text(
    x_inicio_flecha - 0.09, y_inicio_flecha - 0.115,
    "promesas",
    transform=ax.transAxes,
    ha="left",
    va="bottom",
    fontsize=11,
    fontweight="700",
    color=COL_GOLD,
    clip_on=False
)

# Estrellita alineada con "promesas" y más grande
ax.text(
    x_inicio_flecha + 0.040, y_inicio_flecha - 0.130,
    "★",
    transform=ax.transAxes,
    ha="left",
    va="bottom",
    fontsize=18,
    fontweight="700",
    color=COL_GOLD,
    clip_on=False
)

# ============================================================
# 8. ESTETICA
# ============================================================
for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

ax.tick_params(axis="both", length=0)

# Espacio para la anotación externa
plt.subplots_adjust(left=0.28, bottom=0.34)

plt.savefig(
    "factor_k_rating_estilo.png",
    dpi=150,
    bbox_inches="tight",
    pad_inches=0.3
)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
# 1. PARAMETROS DEL CUADRADO RESALTADO DEL HEATMAP
# ============================================================
RATING_PROMESA = 2000
K_PROMESA = 40

month_names = {
    1: "Ene", 2: "Feb", 3: "Mar", 4: "Abr",
    5: "May", 6: "Jun", 7: "Jul", 8: "Ago",
    9: "Sep", 10: "Oct", 11: "Nov", 12: "Dic"
}

# ============================================================
# 2. PREPARAR DATA BASE
# ============================================================
df_trayectoria = df[
    (df["rating"].notna()) &
    (df["k"].notna()) &
    (df["fideid"].notna()) &
    (df["month"].notna()) &
    (df["active"].notna())
].copy()

# Por seguridad, un registro por jugador y mes
df_trayectoria = (
    df_trayectoria
    .sort_values(["fideid", "month"])
    .drop_duplicates(subset=["fideid", "month"])
)

primer_mes = int(df_trayectoria["month"].min())
ultimo_mes = int(df_trayectoria["month"].max())

# ============================================================
# 3. RECONSTRUIR EL UNIVERSO DEL HEATMAP
#    Activos en diciembre que ya estaban activos en enero
# ============================================================
jugadores_enero_activos = df_trayectoria.loc[
    (df_trayectoria["month"] == primer_mes) &
    (df_trayectoria["active"] == True),
    "fideid"
].dropna().unique()

df_heatmap_universo = df_trayectoria[
    (df_trayectoria["month"] == ultimo_mes) &
    (df_trayectoria["active"] == True) &
    (df_trayectoria["fideid"].isin(jugadores_enero_activos))
].copy()

# Un registro por jugador en el universo del heatmap
df_heatmap_universo = (
    df_heatmap_universo
    .sort_values(["fideid", "month"])
    .drop_duplicates(subset=["fideid"])
)

jugadores_universo = df_heatmap_universo["fideid"].unique()

# ============================================================
# 4. FILTRAR EL CUADRADO RESALTADO DEL HEATMAP
#    Rating >= 2000 y K = 40
# ============================================================
jugadores_promesas = df_heatmap_universo.loc[
    (df_heatmap_universo["rating"] >= RATING_PROMESA) &
    (df_heatmap_universo["k"] == K_PROMESA),
    "fideid"
].unique()

# ============================================================
# 5. DEFINIR BASE PARA CALCULAR CAMBIO DESDE EL PRIMER MES
#    Solo jugadores del universo del heatmap
# ============================================================
base = df_trayectoria[
    (df_trayectoria["month"] == primer_mes) &
    (df_trayectoria["active"] == True) &
    (df_trayectoria["fideid"].isin(jugadores_universo))
].copy()

base = (
    base
    .sort_values(["fideid", "month"])
    .drop_duplicates(subset=["fideid"])
)

base["grupo"] = np.where(
    base["fideid"].isin(jugadores_promesas),
    "Promesas",
    "Resto de jugadores"
)

base = base[["fideid", "rating", "grupo"]].rename(
    columns={"rating": "rating_base"}
)

n_universo = len(jugadores_universo)
n_promesas = base[base["grupo"] == "Promesas"]["fideid"].nunique()
n_resto = base[base["grupo"] == "Resto de jugadores"]["fideid"].nunique()

# Se mantiene para pruebas
print(f"Universo heatmap: {n_universo:,} jugadores")
print(f"Promesas del cuadrado resaltado: {n_promesas:,}")
print(f"Resto del universo heatmap: {n_resto:,}")

# ============================================================
# 6. SEGUIR LA TRAYECTORIA DE LOS MISMOS JUGADORES
# ============================================================
df_seg = df_trayectoria[
    df_trayectoria["fideid"].isin(jugadores_universo)
].merge(base, on="fideid", how="inner")

df_seg["cambio_rating"] = df_seg["rating"] - df_seg["rating_base"]

trayectoria = (
    df_seg
    .groupby(["month", "grupo"], observed=True)
    .agg(
        cambio_promedio=("cambio_rating", "mean"),
        rating_promedio=("rating", "mean"),
        jugadores=("fideid", "nunique")
    )
    .reset_index()
)

# ============================================================
# 7. GRAFICO
# ============================================================
fig, ax = plt.subplots(figsize=(13.2, 6.0))

COL_PROMESAS = "#D9A566"   # ambar
COL_RESTO = "#2D4356"      # azul oscuro
COL_TEXTO = "#333333"

colores = {
    "Promesas": COL_PROMESAS,
    "Resto de jugadores": COL_RESTO
}

lineas = {
    "Promesas": "-",
    "Resto de jugadores": "--"
}

marcadores = {
    "Promesas": "o",
    "Resto de jugadores": "s"
}

orden_grupos = ["Promesas", "Resto de jugadores"]

for grupo in orden_grupos:
    data_g = trayectoria[trayectoria["grupo"] == grupo]

    ax.plot(
        data_g["month"],
        data_g["cambio_promedio"],
        marker=marcadores[grupo],
        markersize=5.8,
        linewidth=2.5,
        linestyle=lineas[grupo],
        color=colores[grupo],
        label=grupo
    )

# Línea base: sin cambio respecto al primer mes
ax.axhline(
    0,
    color="#999999",
    linewidth=1,
    linestyle=":",
    alpha=0.95
)

ax.text(
    12.72, 0.15,
    "Sin cambio",
    ha="right",
    va="bottom",
    fontsize=9,
    color="#777777"
)

# ============================================================
# 8. ANOTAR VALOR FINAL
# ============================================================
offset_final = {
    "Promesas": 0.25,
    "Resto de jugadores": -0.15
}

for grupo in orden_grupos:
    data_g = trayectoria[trayectoria["grupo"] == grupo]
    valor_final = data_g.loc[
        data_g["month"] == ultimo_mes,
        "cambio_promedio"
    ].values[0]

    ax.text(
        ultimo_mes + 0.18,
        valor_final + offset_final[grupo],
        f"{valor_final:+.1f} pts.",
        ha="left",
        va="center",
        fontsize=10,
        fontweight="600",
        color=COL_TEXTO,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.9, pad=1.5)
    )

# ============================================================
# 9. FORMATO GENERAL
# ============================================================
ax.set_title(
    "",
    fontsize=14,
    fontweight="600",
    pad=20,
    loc="center",
    color="#222222"
)

ax.set_xlabel("")
ax.set_ylabel("Cambio promedio de nivel de juego", fontsize=11, color=COL_TEXTO)

ax.set_xticks(list(month_names.keys()))
ax.set_xticklabels(list(month_names.values()), fontsize=9)

ax.set_xlim(0.7, 12.9)

y_min = trayectoria["cambio_promedio"].min()
y_max = trayectoria["cambio_promedio"].max()
margen = max(2, (y_max - y_min) * 0.25)
ax.set_ylim(y_min - margen, y_max + margen)

ax.grid(axis="y", alpha=0.32, linestyle="-", linewidth=0.8)
ax.grid(axis="x", alpha=0.08, linestyle="--")
ax.set_axisbelow(True)

for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)

ax.spines["bottom"].set_color("#cccccc")
ax.tick_params(axis="y", length=0, labelsize=9, colors=COL_TEXTO)
ax.tick_params(axis="x", colors=COL_TEXTO)

ax.legend(
    loc="lower center",
    bbox_to_anchor=(0.5, -0.24),
    ncol=2,
    frameon=False,
    fontsize=10
)

ax.text(
    0.5,
    -0.30,
    "La línea muestra cambio promedio respecto al primer mes.",
    transform=ax.transAxes,
    ha="center",
    va="top",
    fontsize=9.1,
    color="#777777",
    style="italic"
)

# Ajuste manual para que no se corte el eje Y, leyenda ni nota
fig.subplots_adjust(left=0.12, right=0.97, top=0.93, bottom=0.32)

plt.savefig(
    "trayectoria_rating_promesas_vs_resto.png",
    dpi=150,
    bbox_inches="tight",
    pad_inches=0.3
)
plt.show()
```
